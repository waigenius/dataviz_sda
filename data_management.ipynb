### Gestion des valeurs manquantes

#### Imputation des variables `title_status` et `condition`

On commence par examiner les deux variables qui décrivent l’état du véhicule :

```python
print(df['title_status'].unique())
print(df['condition'].unique())
```

Ces deux colonnes décrivent toutes deux l’état du véhicule, mais title_status est plus objective que condition.

On examine comment ces deux variables se recoupent grâce à un tableau croisé (matrice de contingence) :

```python
contingency_table = pd.crosstab(df['title_status'], df['condition'])
print(contingency_table)
```

La majorité des véhicules avec un title_status = clean sont en good ou excellent état.

À l’inverse, title_status = salvage ou parts only est souvent associé à une condition dégradée.

On peut donc utiliser la valeur de title_status pour estimer raisonnablement les valeurs manquantes dans condition.

```python
# Remplacer les NaN de title_status par 'missing'
df['title_status'] = df['title_status'].fillna('missing')

# Certaines valeurs de title_status laissent peu de doute sur la condition
# On applique une imputation basée sur la logique métier : un véhicule "salvage" est sûrement aussi "salvage" en condition
mapping_condition = {
    'salvage': 'salvage',
    'parts only': 'salvage'
}

for status, cond in mapping_condition.items():
    df.loc[df['condition'].isna() & (df['title_status'] == status), 'condition'] = cond
    
# Pour les cas restants, on choisit d’imputer les valeurs manquantes de condition par la valeur la plus fréquente pour chaque statut
def impute_with_mode(group):
    mode_val = group.mode()
    return group.fillna(mode_val[0] if not mode_val.empty else 'unknown condition')

df['condition'] = df.groupby('title_status')['condition'].transform(impute_with_mode)

# On vérifie qu’il ne reste aucune valeur manquante dans condition, et que la distribution est cohérente.
print(df['condition'].isna().sum())  # Doit renvoyer 0
print(pd.crosstab(df['title_status'], df['condition']))
```

En conclusion, nous avons mené une exploration approfondie des variables `title_status` et `condition`, qui décrivent toutes deux l’état du véhicule. L’analyse a révélé une forte corrélation entre ces deux colonnes, avec une information plus fiable et objective dans `title_status`. Cette observation nous a permis d’adopter une stratégie d’imputation conditionnelle cohérente pour compléter les valeurs manquantes dans `condition`. On a d’abord appliqué une logique métier simple pour certains statuts très parlants comme "salvage" ou "parts only", puis complété les valeurs restantes en utilisant la modalité la plus fréquente pour chaque statut. Cette démarche a permis de restaurer la complétude de la variable `condition` tout en conservant une cohérence interne dans les données, ce qui améliore la qualité globale du jeu de données et garantit des analyses plus fiables par la suite.

#### Imputation des valeurs manquantes dans les variables `manufacturer`, `drive`, `cylinders`, `fuel`, `transmission`, `size` et `type`

Afin de compléter les valeurs manquantes dans certaines colonnes du jeu de données, nous avons mis en place une fonction d’imputation appliquée à chaque combinaison unique de model et d'une colonne cible (par exemple : drive, cylinders, fuel, transmission…). Cette fonction suit une logique en quatre étapes :

Aucun NaN présent dans la série : la série est retournée telle quelle, sans modification.

Toutes les valeurs sont manquantes (que des NaN) : la série est également retournée sans modification, car aucune information ne permet une imputation fiable.

Une seule valeur observée (non nulle) : tous les NaN sont remplacés par cette valeur unique, considérée comme représentative du modèle.

Plusieurs valeurs observées : les NaN sont imputés aléatoirement, en respectant la distribution empirique des valeurs existantes. Cela permet de conserver la variabilité des données tout en remplissant les valeurs manquantes de manière cohérente.

Cette approche permet une imputation contextualisée et réaliste, en s'appuyant sur la cohérence interne des données pour chaque modèle.

```python
# Dans un premier temps, la liste cols identifie les colonnes sur lesquelles l’imputation va porter : 'manufacturer', 'drive', 'cylinders', 'fuel' et 'transmission'
cols = ['manufacturer', 'drive', 'cylinders', 'fuel', 'transmission', 'size' , 'type']

# Une copie du DataFrame original est réalisée sous le nom df_pre, afin de conserver l’état initial des données avant modification
df_pre = df.copy()

# On calcule et affiche le nombre de valeurs manquantes (NaN) pour chacune de ces colonnes
na_counts = df[cols].isna().sum()
print("NaN par colonne :\n", na_counts, "\n")

# Pour comprendre comment procéder à l’imputation, le code examine combien de valeurs différentes existent pour chaque variable (cols) à l’intérieur de chaque modèle de voiture (colonne model). 
# Cela permet d’identifier les cas où tous les enregistrements d’un même modèle partagent une seule et même valeur (cas simple) et ceux où plusieurs valeurs coexistent (cas complexe). 
# Le résultat est résumé dans un DataFrame summary.
uniques = df.groupby('model')[cols].nunique()
summary = pd.DataFrame({
    'single_value_models':  (uniques == 1).sum(),
    'multiple_value_models': (uniques > 1).sum()
})
print("Nombre de modèles par colonne → 1 seule valeur vs plusieurs :\n", summary, "\n")

# Un générateur de nombres aléatoires, initialisé avec une graine fixe (42), est ensuite défini avec np.random.default_rng(42). 
# Cela garantit que les tirages aléatoires réalisés plus tard seront reproductibles
rng = np.random.default_rng(42)

# On calcule pour chaque colonne concernée, le mode global, c’est-à-dire la valeur la plus fréquente dans l’ensemble du DataFrame (hors NaN). Ces modes globaux serviront de solution de secours si aucune information n’est disponible localement.
global_modes = {col: df[col].mode(dropna=True)[0] for col in cols}

# La fonction impute_group(x, col) définit la stratégie d’imputation pour un sous-ensemble de données x, correspondant à une colonne col dans un modèle de voiture donné
def impute_group(x, col):
    """
    Impute les NaN dans une série x (un seul model pour la colonne col).
    - Si aucun NaN           : renvoie x tel quel.
    - Si tout est NaN         : remplit par le mode global.
    - Si une seule valeur    : duplique cette valeur pour tous les NaN.
    - Sinon                   : tirage aléatoire pondéré par la fréquence locale.
    """
    non_na = x.dropna()
    n_miss = x.isna().sum()

    # Si la série ne contient aucune valeur manquante, elle est retournée telle quelle
    if n_miss == 0:
        return x

    if non_na.empty:
        # fallback global si aucun échantillon local
        return x.fillna(global_modes[col])
    # Si une seule valeur non nulle existe, elle est utilisée pour remplacer les NaN
    if non_na.nunique() == 1:
        return x.fillna(non_na.iloc[0])

    # Sinon, on effectue un tirage aléatoire pondéré parmi les valeurs existantes, selon leur fréquence.
    dist = non_na.value_counts(normalize=True)
    fill_vals = rng.choice(dist.index, size=n_miss, p=dist.values)
    x.loc[x.isna()] = fill_vals
    return x

# --- Imputation ---
# Le cœur de l’imputation se fait dans une boucle for sur chaque colonne de la liste cols
for col in cols:
    # Pour chaque colonne, on applique la fonction impute_group groupe par groupe, en fonction de la variable model. Cela constitue l’imputation locale.
    filled = (
        df
        .groupby('model')[col]
        .transform(lambda s: impute_group(s, col))
    )
    # Une fois cette opération réalisée, les valeurs restant éventuellement manquantes (si aucune valeur valide n'était présente dans un groupe) sont remplacées par le mode global.
    filled = filled.fillna(global_modes[col])
    # Enfin, le résultat est affecté à la colonne correspondante du DataFrame.
    df[col] = filled

# À la fin du processus, un dernier affichage vérifie qu’il ne reste plus aucune valeur manquante dans les colonnes traitées
print("NaN restants :", df[cols].isna().sum().to_dict())

# Le DataFrame final, après imputation, est copié sous le nom df_post
df_post = df.copy()
```

Pour vérifier que les propotions sont bien respectées, nous allons d’abord repéré pour chaque variable (manufacturer, drive, cylinders, fuel et transmission) les modèles dont le taux de valeurs manquantes, avant traitement, se situait strictement entre 5 % et 100 %, puis, parmi ces modèles, sélectionner les trois présentant le plus grand nombre de modalités distinctes. Pour chacun de ces modèles, il va calculé la répartition proportionnelle des valeurs avant imputation, puis la même répartition après imputation, en arrondissant chaque proportion à deux décimales. La présentation côte-à-côte des distributions avant et après traitement nous a permis de vérifier que les proportions relatives des différentes modalités étaient conservées et qu’aucun biais notable n’avait été introduit par le processus de remplissage des données manquantes.

```python
cols = ['manufacturer', 'drive', 'cylinders', 'fuel', 'transmission', 'size' , 'type']

for col in cols:
    # a) Taux de NaN et nombre de valeurs distinctes par model AVANT
    grp = df_pre.groupby('model')[col]
    nan_rate = grp.apply(lambda s: s.isna().mean())
    nunique = grp.nunique()

    # b) Filtrer modèles 5% < NaN < 100%
    valid = nunique.index[(nan_rate > 0.05) & (nan_rate < 1.0)]

    # c) Top 3 modèles selon nunique pour cette colonne
    top3 = nunique.loc[valid].nlargest(3).index.tolist()
    print(f"\n=== Colonne `{col}` — Top 3 modèles ===\n{top3}")

    # d) Afficher pour chacun la distribution avant/après, arrondie à 2 décimales
    for model in top3:
        dist_pre = (
            df_pre.loc[df_pre['model'] == model, col]
                  .value_counts(normalize=True)
        )
        dist_post = (
            df_post.loc[df_post['model'] == model, col]
                   .value_counts(normalize=True)
        )
        comp = pd.DataFrame({'avant': dist_pre, 'après': dist_post}) \
                 .fillna(0) \
                 .sort_index() \
                 .round(2)

        print(f"\nModèle `{model}` (NaN avant : {nan_rate[model]:.2f}, valeurs distinctes : {nunique[model]})")
        print(comp)
```

### Gestion des doublons

#### Premier dédoublonnage en utilisant l'immatriculation du véhicule `VIN` et la date du posting `posting_date`

Ce traitement vise à supprimer les doublons présents dans le DataFrame `df`, mais uniquement pour les lignes contenant un **numéro VIN** (*Vehicle Identification Number*). Le numéro VIN permet d’identifier un véhicule de manière unique, ce qui est essentiel pour détecter les annonces en doublon.

Étape 1 : Isolation des lignes avec un VIN valide

On commence par créer un sous-ensemble `df_vin_non_null`, qui contient uniquement les lignes où la colonne `VIN` est renseignée (c’est-à-dire différente de vide (`''`) et de valeur manquante (`NaN`)). Ces lignes représentent les véhicules identifiables de manière unique.

Étape 2 : Suppression des doublons sur VIN et date de publication

Une fois ce sous-ensemble isolé, on applique la méthode `drop_duplicates()` en se basant sur les colonnes `VIN` et `posting_date`. Cela permet de supprimer les annonces répétées pour un même véhicule publiées à la même date. Le résultat est stocké dans le DataFrame `df_vin_dedoublonne`.

Étape 3 : Conservation des lignes sans VIN

En parallèle, on extrait les lignes qui **ne possèdent pas de VIN valide**. Ces entrées sont conservées dans un second sous-ensemble `df_sans_vin`, car bien qu’elles ne puissent pas être dédupliquées sur la base du VIN, elles peuvent rester utiles pour certaines analyses.

Étape 4 : Fusion des deux sous-ensembles

Enfin, on combine les deux DataFrames (`df_vin_dedoublonne` et `df_sans_vin`) en un seul DataFrame final grâce à `pd.concat()`, avec l’option `ignore_index=True` pour réinitialiser les index. Le DataFrame `df` contient ainsi toutes les données originales, avec les doublons supprimés uniquement là où cela était possible et pertinent.

```python
# Exclure les lignes sans VIN (on ne veut pas les traiter comme doublons)
# On recupère les voitures avec Matricules
df_vin_non_null = df[df['VIN'].notna() & (df['VIN'] != '')]

# Dédoublonnage sur (VIN + posting_date) parmi les lignes valides
df_vin_dedoublonne = df_vin_non_null.drop_duplicates(subset=['VIN', 'posting_date'])

# Garder les autres lignes non concernées (avec VIN vide ou NaN)
df_sans_vin = df[df['VIN'].isna() | (df['VIN'] == '')]

# Recombiner les deux
df = pd.concat([df_vin_dedoublonne, df_sans_vin], ignore_index=True)

# === Résumé du dédoublonnage ===
n_total_avant = df_vin_non_null.shape[0]
n_total_apres = df_vin_dedoublonne.shape[0]
n_doublons = n_total_avant - n_total_apres
pct_doublons = (n_doublons / n_total_avant) * 100
n_sans_vin = df_sans_vin.shape[0]
n_total_final = df.shape[0]

print("\n=== Résumé du dédoublonnage ===")
print(f"Total initial de lignes avec VIN : {n_total_avant}")
print(f"Nombre de doublons écartés (même VIN + posting_date) : {n_doublons} ({pct_doublons:.2f}%)")
print(f"Total de lignes conservées avec VIN : {n_total_apres}")
print(f"Total de lignes sans VIN (non traitées) : {n_sans_vin}")
print(f"Total final après recombinaison : {n_total_final}")

# Supprimée VIN après le dédoublonnage, car elle n'est plus nécessaire pour l’analyse.
df = df.drop(columns='VIN')
```

#### Deuxième dédoublonnage sur les lignes sans 'VIN' en utilisant les variables importantes et la date du posting 'posting_date'

Après avoir effectué un premier nettoyage basé sur le numéro VIN, nous réalisons un second dédoublonnage plus général, basé cette fois sur des **attributs clés** qui décrivent le véhicule de manière précise.

L’objectif est de repérer et supprimer les doublons pour lesquels le véhicule présente exactement les mêmes valeurs sur plusieurs colonnes caractéristiques. Ce type de doublon peut apparaître lorsqu'une même annonce est saisie plusieurs fois avec des identifiants manquants ou différents, mais des attributs identiques.

Nous avons retenu les colonnes suivantes comme représentatives d’un véhicule unique dans le contexte d’une annonce :

- **`price`** : le prix de vente proposé
- **`year`** : l’année du véhicule
- **`manufacturer`** : le constructeur
- **`model`** : le modèle
- **`posting_date`** : la date de publication de l’annonce

La méthode `drop_duplicates()` est appliquée en se basant sur ces colonnes. Seules les lignes présentant une combinaison strictement identique de ces valeurs sont considérées comme des doublons et supprimées.
Ce dédoublonnage complémentaire permet d’assurer une meilleure qualité des données en éliminant les répétitions involontaires qui ne peuvent pas être détectées via le numéro VIN.


```python
# Définir les colonnes clés pour le second dédoublonnage
colonnes_importantes = ['price', 'year', 'manufacturer', 'model', 'posting_date']
n_avant = df.shape[0]

# Appliquer le drop_duplicates
df = df.drop_duplicates(subset=colonnes_importantes)
n_apres = df.shape[0]

# === Résumé du dédoublonnage ===
n_doublons = n_avant - n_apres
pct_doublons = (n_doublons / n_avant) * 100

print("\n=== Résumé du dédoublonnage (par attributs clés) ===")
print(f"Total initial de lignes : {n_avant}")
print(f"Nombre de doublons écartés (même combinaison colonnes importantes) : {n_doublons} ({pct_doublons:.2f}%)")
print(f"Total final après dédoublonnage : {n_apres}")
```

### Traitement des valeurs abérrantes (outlet)

#### Gestion des valeurs aberrantes – Variable `odometer`

On utilise ce code pour analyser la répartition des valeurs extrêmes de odometer, afin d’identifier des seuils pertinents pour le nettoyage des données avant toute analyse.

```python
# Total de lignes pour référence
total_lignes = df.shape[0]

# Seuils inférieurs
seuils_inferieurs = [0, 10, 50, 100, 500, 1000]
print("=== Nombre de lignes avec odometer inférieur ou égal aux seuils ===")
for seuil in seuils_inferieurs:
    count = (df['odometer'] <= seuil).sum()
    print(f"  ≤ {seuil} miles : {count} lignes ({(count/total_lignes)*100:.2f}%)")

# Seuils supérieurs
seuils_superieurs = [150000, 200000, 300000, 400000, 500000]
print("\n=== Nombre de lignes avec odometer supérieur aux seuils ===")
for seuil in seuils_superieurs:
    count = (df['odometer'] > seuil).sum()
    print(f"  > {seuil} miles : {count} lignes ({(count/total_lignes)*100:.2f}%)")

```

Après analyse de la distribution des valeurs et en croisant les recommandations issues de sources métier, nous avons choisi de restreindre la variable `odometer` aux valeurs comprises entre 100 miles et 300 000 miles**.

Justification du seuil bas (100 miles)

Les véhicules affichant un kilométrage très faible sont probablement mal renseignés ou mal catégorisés (véhicules neufs enregistrés comme voitures d'occasion, erreurs de saisie, etc.).

Dans le dataset, on observe :
- 1 209 lignes avec un kilométrage ≤ 10 miles  
- 3 719 lignes ≤ 50 miles  
- 4 292 lignes ≤ 100 miles

Ces lignes représentent **moins de 1,5 % de l’ensemble du dataset** (371 031 lignes). Leur suppression permet de réduire le bruit, d’éviter des valeurs aberrantes et d’améliorer la lisibilité des visualisations, sans compromettre la qualité de l’échantillon analysé.

Justification du seuil haut (300 000 miles)

Selon la source métier [Mercier Auto]( https://www.mercier-auto.com/guide-achat-auto/combien-de-kilometrage-pour-une-voiture-doccasion/#:~:text=Il%20peut%20s'av%C3%A9rer%20difficile,s'%C3%A9l%C3%A8ve%20%C3%A0%20150.000%20kilom%C3%A8tres.), la revente devient difficile à partir de :
- 150 000 km (soit environ 93 000 miles) pour les véhicules essence,
- 150 000 miles pour les véhicules diesel.

Si on appliquait ces seuils strictement, on exclurait :
- 64 854 lignes pour les véhicules entre 150 000 et 200 000 miles 
- 20 086 lignes supplémentaires entre 200 000 et 300 000 miles

Cela représenterait **près de 23 % du dataset**, ce qui est **excessif** par rapport à l’objectif initial de nettoyage.  
En retenant une limite haute à **300 000 miles**, on élimine les valeurs les plus extrêmes et les cas potentiellement erronés, tout en conservant les données réalistes pour des usages professionnels (taxis, utilitaires, flottes longue distance...).

Conclusion

Le choix de filtrer `odometer` entre **100 et 300 000 miles** repose donc sur trois critères :
- Une **analyse quantitative** des données disponibles
- Une **référence métier** crédible sur les seuils de revente réalistes
- Une **volonté de préserver la lisibilité des graphiques** et la représentativité de l’échantillon

Ce nettoyage ciblé permet d’améliorer la qualité globale du dataset sans perte significative d’information.

Nous avons établi deux séries de seuils :

- **Seuils bas** : `[0, 100, 500, 1000, 2000]`  
  Ces seuils permettent d’identifier les valeurs anormalement faibles, voire incohérentes dans le contexte d’une vente (ex. : prix = 0 ou 1).

```python
total_lignes = df.shape[0]

seuils_bas = [0, 100, 500, 1000, 2000]
seuils_haut = [50000, 100000, 150000, 200000, 500000]

print("=== Lignes avec price ≤ seuils ===")
for seuil in seuils_bas:
    count = (df['price'] <= seuil).sum()
    print(f"  ≤ {seuil} : {count} lignes ({(count/total_lignes)*100:.2f}%)")

print("\n=== Lignes avec price ≥ seuils ===")
for seuil in seuils_haut:
    count = (df['price'] >= seuil).sum()
    print(f"  ≥ {seuil} : {count} lignes ({(count/total_lignes)*100:.2f}%)")
```

À l’inverse, on observe également des valeurs très élevées :

- 7 842 lignes (2.70%) ont un prix **supérieur ou égal à 50 000**.
