## Nettoyage

```python
df = pd.read_csv("vehicles.csv")
```

### Suppression des colonnes inutiles

```python
df.drop(columns=['url','region_url','image_url','description'], inplace=True)
```

```python
df.columns
```

### Traitement des valeurs incohérentes

```python
#Suppression des prix incohérents
df = df[(df['price'] >= 1000) & (df['price'] <= 250000)]
df.shape
```

```python
#Suppression des années incohérentes
df = df[(df['year'] >= 1950) & (df['year'] <= 2025)]
df.shape
```

```python
#Suppression des kilométrages incohérents
df = df[(df['odometer'] >= 100) & (df['odometer'] <= 500000)]
df.shape
```

### Gestion des valeurs manquantes

```python
#Suppression des lignes avec des valeurs manquantes importantes
colonnes_importantes = ['price', 'year', 'manufacturer', 'model', 'odometer']
df.dropna(subset=colonnes_importantes, inplace=True)
df.shape
```

### Correction des types de données

```python
# Conversion de la colonne 'posting_date' en format datetime sans les heures
df['posting_date'] = pd.to_datetime(df['posting_date'], errors='coerce', utc=True)
df['posting_date'] = df['posting_date'].dt.tz_convert(None).dt.normalize()
```

```python
df.sample(n=10)
```

### Gestion des doublons

#### Premier dédoublonnage en utilisant l'immatriculation du véhicule 'VIN' et la date du posting 'posting_date'

```python
df.shape
```

```python
# Exclure les lignes sans VIN (on ne veut pas les traiter comme doublons)
df_vin_non_null = df[df['VIN'].notna() & (df['VIN'] != '')]
df_vin_non_null.shape
```

```python
# Dédoublonnage sur (VIN + posting_date) parmi les lignes valides
df_vin_dedoublonne = df_vin_non_null.drop_duplicates(subset=['VIN', 'posting_date'])
df_vin_dedoublonne.shape
```

```python
# Garder les autres lignes non concernées (avec VIN vide ou NaN)
df_sans_vin = df[df['VIN'].isna() | (df['VIN'] == '')]
df_sans_vin.shape
```

```python
# Recombiner les deux
df = pd.concat([df_vin_dedoublonne, df_sans_vin], ignore_index=True)
df.shape
```

#### Deuxième dédoublonnage sur les lignes sans 'VIN' en utilisant les variables importantes et la date du posting 'posting_date'

```python
# Définir les colonnes clés pour le second dédoublonnage
colonnes_importantes = [
    'region', 'price', 'year', 'manufacturer', 'model', 'posting_date'
]
df.shape
```

```python
# Appliquer le drop_duplicates
df = df.drop_duplicates(subset=colonnes_importantes)
df.shape
```

Conclusion du dédoublonnage :
- 77 017 doublons supprimés en utilisant les colonnes 'VIN' + 'Posting_date' (~35%), ce qui montre un très fort taux de repost d'annonce des mêmes voitures dans la même journée
- 1 494 doublons supprimés sur en utilisant les colonnes importantes (~1%)
Cette suppression évite d'avoir des annonces identiques ou trop proches pour augmenter la qualité de nos analyses
